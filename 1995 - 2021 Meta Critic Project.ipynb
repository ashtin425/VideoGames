```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
from matplotlib import pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts the configuration of plots we will create
```

```python
# imporing csv file
vg = pd.read_csv('all_games.csv')
vg.head()
```

```python
# looking at each unique platform in dataset
vg.platform.unique()
```

```python
# looking at potential null values
vg.info()
```

```python
# creating variables for each console platform

N_64 = vg[vg.platform == 'Nintendo64']
PlayStation = vg[vg.platform == 'PlayStation']
PlayStation3 = vg[vg.platform == 'PlayStation3']
Dreamcast = vg[vg.platform == 'Dreamcast']
Xbox360 = vg[vg.platform == 'Xbox360']
Wii = vg[vg.platform == 'Wii']
XboxOne = vg[vg.platform == 'XboxOne']
PC = vg[vg.platform == 'PC']
Switch = vg[vg.platform == 'Switch']
PlayStation2 = vg[vg.platform == 'PlayStation2']
PlayStation4 = vg[vg.platform == 'PlayStation4']
GameCube = vg[vg.platform == 'GameCube']
Xbox = vg[vg.platform == 'Xbox']
WiiU = vg[vg.platform == 'WiiU']
GameBoyAdvance = vg[vg.platform == 'GameBoyAdvance']
DS = vg[vg.platform == '3DS']
PlayStationVita = vg[vg.platform == 'PlayStationVita']
PlayStation5 = vg[vg.platform == 'PlayStation5']
PSP = vg[vg.platform == 'PSP']
Stadia = vg[vg.platform == 'Stadia']
```

```python
# sorting each platform by top meta scores
Meta_N_64 = N_64.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStation = PlayStation.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStation3 = PlayStation3.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Dreamcast = Dreamcast.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Xbox360 = Xbox360.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Wii = Wii.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_XboxOne = XboxOne.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PC = PC.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Switch = Switch.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStation2 = PlayStation2.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStation4 = PlayStation4.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_GameCube = GameCube.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Xbox = Xbox.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_WiiU = WiiU.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_GameBoyAdvance = GameBoyAdvance.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_DS = DS.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStationVita = PlayStationVita.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PlayStation5 = PlayStation5.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_PSP = PSP.sort_values(by = ['meta_score'], inplace = False, ascending = False)
Meta_Stadia = PlayStation.sort_values(by = ['meta_score'], inplace = False, ascending = False)
```
